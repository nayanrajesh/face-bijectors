```python
def RidgeIso(P,Q):
    Prel = [x for x in P.relations() if x[0]!=0 and x[0]!=x[1]]
    Pel = list(set([x[0] for x in Prel]))
    Qrel = [x for x in Q.relations() if x[0]!=0 and x[0]!=x[1]]
    Qel = list(set([x[0] for x in Qrel]))
    if len(Pel)==len(Qel):
        return True
    return False
```
